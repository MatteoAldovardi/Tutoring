# Le Coniche: Un Viaggio tra Curve Speciali

Le coniche sono curve affascinanti che compaiono in molti ambiti, dalla matematica alla fisica, all'astronomia. In questa lezione, esploreremo le coniche dal punto di vista della geometria analitica, concentrandoci sulle loro definizioni come luoghi geometrici.

## Cosa sono le Coniche?

Le coniche sono curve ottenute dall'intersezione di un piano con un cono a due falde. A seconda dell'inclinazione del piano rispetto al cono, otteniamo diverse tipologie di coniche:

* **Circonferenza:** il piano interseca il cono perpendicolarmente all'asse.
* **Ellisse:** il piano interseca il cono obliquamente, tagliando entrambe le falde.
* **Parabola:** il piano interseca il cono parallelamente a una generatrice.
* **Iperbole:** il piano interseca entrambe le falde del cono.

## Definizioni di Luogo Geometrico

Un luogo geometrico è un insieme di punti che soddisfano una determinata proprietà. Le coniche possono essere definite come luoghi geometrici in diversi modi, utilizzando punti speciali chiamati fuochi e una retta speciale chiamata direttrice.

### 1. Ellisse

L'ellisse è il luogo geometrico dei punti del piano tali che la somma delle distanze da due punti fissi, detti fuochi (F1 e F2), è costante.

* **Fuochi:** $F_1$ e $F_2$
* **Costante:** $2a$ (dove $a$ è il semiasse maggiore)



```python
%matplotlib widget  


import matplotlib.pyplot as plt
import numpy as np
from matplotlib.widgets import Slider


# Parametri iniziali
a_init = 5
b_init = 3

# Crea la figura e gli assi
fig, ax = plt.subplots()
plt.subplots_adjust(left=0.1, bottom=0.3)  # Aumentato spazio per sliders

# Plot iniziale dell'ellisse
x = np.linspace(-a_init, a_init, 400)
y = b_init * np.sqrt(1 - (x/a_init)**2)
line_top, = ax.plot(x, y, 'r', label='Ellisse')
line_bottom, = ax.plot(x, -y, 'r')

# Plot dei fuochi
c_init = np.sqrt(a_init**2 - b_init**2)
f1, = ax.plot(-c_init, 0, 'bo', label='Fuochi')
f2, = ax.plot(c_init, 0, 'bo')

# Segmenti dai fuochi a un punto sull'ellisse
point_x = a_init / 2
point_y = b_init * np.sqrt(1 - (point_x/a_init)**2)
segment1, = ax.plot([-c_init, point_x], [0, point_y], 'g-')
segment2, = ax.plot([c_init, point_x], [0, point_y], 'g-')
point, = ax.plot(point_x, point_y, 'go')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Ellisse con parametri regolabili')
ax.legend()
ax.grid(True)
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.set_aspect('equal', adjustable='box')

# Slider per il parametro a
ax_a = plt.axes([0.1, 0.15, 0.8, 0.03])
slider_a = Slider(ax_a, 'a', 1, 10, valinit=a_init)

# Slider per il parametro b
ax_b = plt.axes([0.1, 0.1, 0.8, 0.03])
slider_b = Slider(ax_b, 'b', 1, a_init, valinit=b_init)

# Funzione per aggiornare il plot
def update(val):
    a = slider_a.val
    b = slider_b.val

    # Impedisce b > a
    if b > a:
        b = a
        slider_b.set_val(b)  # Modifica b senza generare un loop di eventi

    c = np.sqrt(a**2 - b**2)

    x = np.linspace(-a, a, 400)
    y = b * np.sqrt(1 - (x/a)**2)
    line_top.set_xdata(x)
    line_top.set_ydata(y)
    line_bottom.set_xdata(x)
    line_bottom.set_ydata(-y)

    f1.set_xdata(-c)
    f2.set_xdata(c)

    point_x = a / 2
    point_y = b * np.sqrt(1 - (point_x/a)**2)
    segment1.set_xdata([-c, point_x])
    segment1.set_ydata([0, point_y])
    segment2.set_xdata([c, point_x])
    segment2.set_ydata([0, point_y])
    point.set_xdata(point_x)
    point.set_ydata(point_y)

    fig.canvas.draw_idle()

# Collega gli slider alla funzione di aggiornamento
slider_a.on_changed(update)
slider_b.on_changed(update)

plt.show()
```

```python
### 2. Parabola

La parabola è il luogo geometrico dei punti del piano equidistanti da un punto fisso, detto fuoco ($F$), e da una retta fissa, detta direttrice ($d$).

* **Fuoco:** $F$
* **Direttrice:** $d$

### 3. Iperbole

L'iperbole è il luogo geometrico dei punti del piano tali che il valore assoluto della differenza delle distanze da due punti fissi, detti fuochi ($F_1$ e $F_2$), è costante.

* **Fuochi:** $F_1$ e $F_2$
* **Costante:** $2a$ (dove $a$ è il semiasse trasverso)

## Equazioni delle Coniche

Le coniche possono essere descritte da equazioni matematiche. Le equazioni più comuni sono:

* **Circonferenza:** $$(x - h)² + (y - k)² = r²$$
* **Ellisse:** $$(x²/a²) + (y²/b²) = 1 $$
* **Parabola:** $$ y = ax² + bx + c $$
* **Iperbole:** $$ (x²/a²) - (y²/b²) = 1 $$
```

```python


# Iperbole
x = np.linspace(-2*a, 2*a, 400)
y = b * np.sqrt((x/a)**2 - 1)
plt.plot(x, y, 'b', label='Iperbole')
plt.plot(x, -y, 'b')

# Parabola
x = np.linspace(-10, 10, 400)
y = x**2 / (4*a)  # Esempio di parabola
plt.plot(x, y, 'g', label='Parabola')

# Circonferenza
theta = np.linspace(0, 2*np.pi, 400)
x = a * np.cos(theta)
y = a * np.sin(theta)
plt.plot(x, y, 'y', label='Circonferenza')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Coniche')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
