# Lezione sui Poligoni Inscritti e Circoscritti

## Introduzione
I poligoni inscritti e circoscritti sono figure geometriche che hanno una relazione speciale con una circonferenza.  
**Obiettivo della lezione**: Comprendere definizioni, proprietà e applicazioni di questi poligoni.

---

## 1. Poligoni Inscritti (Ciclici)

### 1.1 Definizione
- Un poligono si dice **inscritto** se tutti i suoi vertici giacciono su una circonferenza, detta **circonferenza circoscritta**.
- Esempio classico: il triangolo rettangolo inscritto in una semicirconferenza.

### 1.2 Proprietà Fondamentali
- **Triangoli**: Tutti i triangoli sono inscrivibili in una circonferenza.  
  Il centro della circonferenza circoscritta (circocentro) è l'intersezione degli assi dei lati.
- **Quadrilateri ciclici**: Un quadrilatero è inscrivibile se e solo se la somma degli angoli opposti è 180°.  
  Formula angoli: $ \alpha + \gamma = \beta + \delta = 180^\circ $.

### 1.3 Teoremi Importanti
- **Teorema di Tolomeo**: In un quadrilatero ciclico, il prodotto delle diagonali è uguale alla somma dei prodotti dei lati opposti:  
  $AC \cdot BD = AB \cdot CD + BC \cdot DA $.
- **Raggio della circonferenza circoscritta** (per triangoli):  
  $$ R = \frac{a \cdot b \cdot c}{4 \cdot \text{Area}} $$.

### 1.4 Esempi
- **Esempio 1**: Un rettangolo è sempre ciclico (i vertici giacciono su una circonferenza).  
- **Esempio 2**: Un trapezio isoscele è ciclico se e solo se è rettangolo.

---

## 2. Poligoni Circoscritti (Tangenziali)

### 2.1 Definizione
- Un poligono si dice **circoscritto** se tutti i suoi lati sono tangenti a una circonferenza, detta **circonferenza inscritta**.
- Esempio: un rombo, dove il centro della circonferenza è l'intersezione delle diagonali.

### 2.2 Proprietà Fondamentali
- **Triangoli**: Tutti i triangoli hanno una circonferenza inscritta.  
  Il centro (incentro) è l'intersezione delle bisettrici degli angoli.
- **Quadrilateri tangenziali**: Un quadrilatero è circoscrivibile se e solo se la somma dei lati opposti è uguale:  
  \( AB + CD = BC + DA \).

### 2.3 Teoremi Importanti
- **Area di un poligono tangenziale**:  
  \( \text{Area} = r \cdot s \), dove \( r \) è il raggio della circonferenza inscritta e \( s \) è il semiperimetro.
- **Raggio della circonferenza inscritta** (per triangoli):  
  \( r = \frac{\text{Area}}{s} \).

### 2.4 Esempi
- **Esempio 1**: Un quadrato è sia ciclico che tangenziale.  
- **Esempio 2**: Un deltoide è tangenziale solo se è un rombo.

---

## 3. Confronto tra Poligoni Inscritti e Circoscritti

| Caratteristica          | Poligono Inscritto           | Poligono Circoscritto       |
|-------------------------|-------------------------------|------------------------------|
| **Definizione**         | Vertici su una circonferenza  | Lati tangenti a una circonferenza |
| **Centro**              | Circocentro                   | Incentro                     |
| **Condizioni necessarie** | Angoli opposti supplementari (quadrilateri) | Somma lati opposti uguale (quadrilateri) |
| **Esempi comuni**       | Rettangolo, trapezio isoscele | Rombo, quadrato              |

---

## 4. Applicazioni Pratiche
- **Architettura**: Cupole circolari con elementi inscritti.
- **Ingegneria**: Ruote dentate circoscritte per trasmissioni meccaniche.
- **Problemi classici**: Calcolo di aree massime/minime con vincoli geometrici.

## 5. Esercizi Guidati

### Esercizio 1 (Inscritto)
Un quadrilatero ha angoli \( 70^\circ, 110^\circ, 80^\circ, 100^\circ \). È ciclico?  
**Soluzione**:  
$$ 70^\circ + 110^\circ = 180^\circ $$ e $$ 80^\circ + 100^\circ = 180^\circ $$
Di conseguenza, il quadrilatero è inscrivibile.



### Esercizio 2 (Circoscritto)
Calcola il raggio della circonferenza inscritta in un triangolo con lati 6 cm, 8 cm, 10 cm.  
**Soluzione**:  
- Semipermimetro: $ s = \frac{6+8+10}{2} = 12 $  
- Area (triangolo rettangolo): $ \frac{6 \cdot 8}{2} = 24 $  
- Raggio: $ r = \frac{24}{12} = 2 $a cm.
