### **Study of Functions**
1. $$ f(x) = e^x - x^2 $$
   - **Domain**: Find the domain of definition.  
   - **First derivative**: Compute \( f'(x) \) and find critical points.  
   - Identify local minima/maxima.  
   - **Concavity & Inflection Points:** Compute \( f''(x) \) and analyze concavity.  

2. $$ f(x) = x e^{-x} $$
   - **Domain & Range:** Identify the function's domain and range.
   - **First Derivative & Critical Points:** Compute \( f'(x) \) and find critical points.
   - **Second Derivative & Concavity:** Compute \( f''(x) \) and find inflection points.
   - **Asymptotes & Limits:** Study behavior as \( x \to \pm \infty \).

3. $$ f(x) = \ln x - e^{-x} $$  
   - **Domain:** Find the domain of definition.  
   - **First Derivative & Monotonicity:** Compute \( f'(x) \) and determine where the function is increasing/decreasing.  
   - **Second Derivative & Concavity:** Find \( f''(x) \) to determine intervals of concavity.
   - **Asymptotic Behavior:** Find the limits as $$ x \to 0^+ \) and \( x \to \infty $$.
