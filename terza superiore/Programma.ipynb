## Geometria Analitica del Piano

* **Il piano cartesiano:**
    * Definizione e rappresentazione grafica.
* **Le coordinate di un punto nel piano:**
    * Ascissa e ordinata.
* **I quadranti:**
    * Definizione e individuazione dei punti nei quadranti.
* **La distanza tra due punti:**
    * Formula della distanza nel caso generale e nel caso di punti allineati.
* **Punto medio di un segmento:**
    * Calcolo delle coordinate del punto medio.
* **Baricentro di un triangolo:**
    * Calcolo delle coordinate del baricentro.
* **Equazione di una retta in forma implicita:**
    * ax + by + c = 0.
* **Rette parallele agli assi:**
    * Equazioni delle rette orizzontali e verticali.
* **Il coefficiente angolare:**
    * Definizione e significato geometrico.
* **Equazione di una retta in forma esplicita:**
    * y = mx + q.
* **Equazione di una retta passante per un punto e di coefficiente angolare noto:**
    * y - y0 = m(x - x0).
* **Coefficiente angolare note le coordinate di due punti:**
    * m = (y2 - y1) / (x2 - x1).
* **Equazione di una retta passante per due punti:**
    * (y - y1) / (y2 - y1) = (x - x1) / (x2 - x1).
* **Posizione reciproca tra rette:**
    * Rette parallele, perpendicolari, incidenti.
* **Condizione di parallelismo e perpendicolarità:**
    * m1=m2 , m1\*m2=-1
* **Distanza di un punto di una retta:**
    * Formula della distanza punto retta.
* **Asse di un segmento:**
    * Definizione e calcolo dell'equazione.
* **Fascio proprio e fascio improprio:**
    * Definizione e rappresentazione grafica.
* **Definizione di parabola:**
    * Luogo geometrico dei punti equidistanti dal fuoco e dalla direttrice.
* **Parabola con asse coincidente con l’asse delle ordinate e vertice nell’origine:**
    * Equazione y = ax^2.
* **Parabola con asse parallelo all’asse delle ordinate:**
    * Equazione y = ax^2 + bx + c.
* **Casi particolari:**
    * Parabola con vertice sull'asse x o y.
* **Parabola con asse parallelo all’asse delle ascisse:**
    * Equazione x = ay^2 + by + c.
* **Rette e parabole:**
    * Intersezione tra retta e parabola, rette secanti, tangenti, esterne.
* **Rette tangenti a una parabola:**
    * Calcolo dell'equazione delle rette tangenti.
* **Cenni sulla circonferenza:**
    * Definizione ed equazione della circonferenza.
