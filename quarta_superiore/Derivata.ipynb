# Notebook sulle Derivate

## 1. Introduzione
La **derivata** di una funzione misura quanto rapidamente la funzione cambia al variare del suo input. È un concetto fondamentale nel calcolo differenziale con applicazioni in fisica, economia e machine learning.

## 2. Definizione Formale
La derivata di $f(x)$ nel punto $x_0$ è:
$$
f'(x_0) = \lim_{h \to 0} \frac{f(x_0 + h) - f(x_0)}{h}
$$
Se il limite esiste, $f$ è **derivabile** in $x_0$.

## 3. Significato Geometrico
La derivata rappresenta il coefficiente angolare della **retta tangente** al grafico della funzione nel punto $x_0$.


```python
import numpy as np
import matplotlib.pyplot as plt

# Esempio: Retta tangente a f(x) = x^2 in x=1
x = np.linspace(-2, 2, 100)
f = lambda x: x**2
df = lambda x: 2*x  # Derivata

x0 = 1
tangente = f(x0) + df(x0)*(x - x0)

plt.plot(x, f(x), label='f(x) = x²')
plt.plot(x, tangente, label='Retta tangente in x=1')
plt.scatter(x0, f(x0), color='red')
plt.legend()
plt.grid(True)
plt.show()
```

# Derivate: Perché i Piccoli Comandi Funzionano?  
## Come un'auto risponde agli aggiustamenti del volante

---

## 1. Analogia Pratica: Controllo di un Veicolo  
Immagina:  
- **$\theta$** = angolo del volante (es: 30°)  
- **$f(\theta)$** = direzione effettiva del veicolo  
- **$\delta$** = piccolo aggiustamento (es: 1°)

**Domanda chiave**:  
*Quanto cambia realmente la direzione quando faccio un micro-aggiustamento $\delta$?*

**Risposta derivata**:  
La variazione stimata è circa:  
$$
\Delta f \approx f'(\theta) \cdot \delta
$$


una derivata prima grande in valore assoluto significa avere un grosso combiamento 



## 2. Visualizzazione Geometrica  
Consideriamo un caso semplice: **$f(x) = x^2$** (modello semplificato di risposta non lineare)

```python
import numpy as np
import matplotlib.pyplot as plt

x0 = 1
f = lambda x: x**2
df = lambda x: 2*x  # Derivata analitica

# Dati per plot
x = np.linspace(0.5, 1.5, 100)
y = f(x)
tangent = f(x0) + df(x0)*(x - x0)

# Plot
plt.figure(figsize=(10,6))
plt.plot(x, y, lw=3, label='f(x) = x²')
plt.plot(x, tangent, '--', lw=2, label='Approssimazione lineare')
plt.scatter(x0, f(x0), color='black', s=100, zorder=5)

# Esempio con δ=0.3
delta = 0.3
plt.annotate('', xy=(x0+delta, f(x0+delta)), xytext=(x0+delta, tangent[np.argmin(abs(x - (x0+delta)))]),
              arrowprops=dict(
                 arrowstyle="->", 
                 color='red',
                 lw=2,               # Spessore linea
                 mutation_scale=10   # Dimensione punta freccia
             ))

plt.annotate('f(x + \delta) - f(x)', # come si piazza sta cosa?)

plt.title("Approssimazione Lineare vs Variazione Reale")
plt.xlabel("x (angolo volante simbolico)")
plt.ylabel("f(x) (risposta direzione)")
plt.legend()
plt.grid(True)
plt.show()
```
