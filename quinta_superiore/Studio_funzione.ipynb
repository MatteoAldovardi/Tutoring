**Studio di funzione: $f(x) = \frac{x^2 - 4}{x - 1}$**

<details>
<summary>Studio di funzione: $f(x) = \frac{x^2 - 4}{x - 1}$</summary>

**1. Dominio:**

$$D = \{x \in \mathbb{R} \mid x \neq 1\}$$

**2. Segno della funzione:**

* **Zeri:**
    * $x^2 - 4 = 0 \Rightarrow x = \pm 2$
* **Segno:**
    * $f(x) > 0$ quando $(x < -2) \cup (1 < x < 2) \cup (x > 2)$
    * $f(x) < 0$ quando $(-2 < x < 1) \cup (x < -2)$

**3. Asintoti:**

* **Asintoto verticale:**
    * $x = 1$
    * $\lim_{x \to 1^-} f(x) = +\infty$
    * $\lim_{x \to 1^+} f(x) = -\infty$
* **Asintoto obliquo:**
    * $m = \lim_{x \to \infty} \frac{f(x)}{x} = 1$
    * $q = \lim_{x \to \infty} (f(x) - mx) = 1$
    * $y = x + 1$

**4. Derivata prima e punti stazionari:**

* **Derivata prima:**
    * $f'(x) = \frac{x^2 - 2x + 4}{(x - 1)^2}$
* **Punti stazionari:**
    * $f'(x) = 0$ non ha soluzioni reali, quindi non ci sono punti stazionari.
* **Monotonia:**
    * $f'(x) > 0$ per ogni $x \in D$, quindi la funzione è sempre crescente.

**5. Derivata seconda, concavità e flessi:**

* **Derivata seconda:**
    * $f''(x) = \frac{-6}{(x - 1)^3}$
* **Concavità:**
    * $f''(x) > 0$ per $x < 1$, quindi la funzione è concava verso l'alto.
    * $f''(x) < 0$ per $x > 1$, quindi la funzione è concava verso il basso.
* **Punti di flesso:**
    * Non ci sono punti di flesso nel dominio della funzione.

</details>

**Esempio 1: Funzione esponenziale**

<details>
  <summary>f(x) = x * e^(-x)</summary>

* **Dominio:** (-∞, +∞)
* **Segno:**
    * f(x) > 0 per x > 0
    * f(x) < 0 per x < 0
    * f(x) = 0 per x = 0
* **Derivata prima:** f'(x) = e^(-x) * (1 - x)
    * f'(x) > 0 per x < 1 (funzione crescente)
    * f'(x) < 0 per x > 1 (funzione decrescente)
    * f'(x) = 0 per x = 1 (punto di massimo)
* **Derivata seconda:** f''(x) = e^(-x) * (x - 2)
    * f''(x) > 0 per x > 2 (concavità verso l'alto)
    * f''(x) < 0 per x < 2 (concavità verso il basso)
    * f''(x) = 0 per x = 2 (punto di flesso)
* **Grafico:** La funzione ha un massimo in x = 1 e un punto di flesso in x = 2. Tende a 0 per x che tende a +∞ e a -∞.
</details>

**Esempio 2: Funzione logaritmica**

<details>
  <summary>f(x) = ln(x) / x</summary>

* **Dominio:** (0, +∞)
* **Segno:**
    * f(x) > 0 per x > 1
    * f(x) < 0 per 0 < x < 1
    * f(x) = 0 per x = 1
* **Derivata prima:** f'(x) = (1 - ln(x)) / x^2
    * f'(x) > 0 per 0 < x < e (funzione crescente)
    * f'(x) < 0 per x > e (funzione decrescente)
    * f'(x) = 0 per x = e (punto di massimo)
* **Derivata seconda:** f''(x) = (2 * ln(x) - 3) / x^3
    * f''(x) > 0 per x > e^(3/2) (concavità verso l'alto)
    * f''(x) < 0 per 0 < x < e^(3/2) (concavità verso il basso)
    * f''(x) = 0 per x = e^(3/2) (punto di flesso)
* **Grafico:** La funzione ha un massimo in x = e e un punto di flesso in x = e^(3/2). Tende a -∞ per x che tende a 0+ e a 0 per x che tende a +∞.
</details>

**Esempio 3: Funzione esponenziale e logaritmica combinata**

<details>
  <summary>f(x) = e^x - ln(x)</summary>

* **Dominio:** (0, +∞)
* **Segno:** f(x) > 0 per tutto il dominio
* **Derivata prima:** f'(x) = e^x - 1/x
    * f'(x) > 0 per tutto il dominio (funzione sempre crescente)
* **Derivata seconda:** f''(x) = e^x + 1/x^2
    * f''(x) > 0 per tutto il dominio (concavità sempre verso l'alto)
* **Grafico:** La funzione è sempre crescente e concava verso l'alto. Tende a +∞ per x che tende a 0+ e a +∞ per x che tende a +∞.
</details>
