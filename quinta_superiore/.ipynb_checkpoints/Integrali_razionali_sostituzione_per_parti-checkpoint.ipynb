# Integrali frazionari:

La funzione razionale
$$
\dfrac{p(t)}{(t - a_1)^{3}(t - a_2)^{2}(t^{2} + B)^{3}} \quad (B > 0, \; a_1 \neq a_2)
$$

può essere decomposta nella forma

$$
q(t) + \dfrac{A}{(t-a_1)^{1}} + \dfrac{B}{(t-a_1)^{2}} + \dfrac{C}{(t-a_1)^{3}} + \dfrac{D}{(t-a_2)^{1}} + \dfrac{E}{(t-a_2)^{2}} \\
+ \dfrac{E \cdot t + F}{(t^2 + B)^{1}} + \dfrac{G \cdot t + H}{(t^2 + B)^{2}},
$$

con $q(t) \neq 0$ se $\deg p(t) > \deg \big((t - a_1)^3 (t-a_2)^2 (t^2 + B)^3\big) = 8.$

Per convincersi è utile questo argomento intuitivo. Immaginate di aver quasi finito il procedimento per ottenere la decomposizione sopra e di essere arrivati alla forma seguente

$$
q(t) + \dfrac{A_1 t + A_2}{(t-a_1)^{1}} + \dfrac{B_1 t + B_2}{(t-a_1)^{2}} + \dfrac{C_1 t + C_2}{(t-a_1)^{3}} + \dfrac{D}{(t-a_2)^{1}} + \dfrac{E}{(t-a_2)^{2}} \\
+ \dfrac{E \cdot t + F}{(t^2 + B)^{1}} + \dfrac{G \cdot t + H}{(t^2 + B)^{2}}.
$$

A questo punto è sufficiente eseguire la divisione per $(t - a_1)$ per i numeratori delle prime tre frazioni sopra!  
Svolgendo i conti si ottiene
$$
q(t) + \dfrac{A_1 \cdot (t-a_1) + a_1 \cdot A_1 + A_2}{(t-a_1)^{1}} + \dfrac{B_1 \cdot (t -a_1) + B_1\cdot a_1 + B_2}{(t-a_1)^{2}} + \dfrac{C_1\cdot (t -a_1) + a_1 \cdot C_1 + C_2}{(t-a_1)^{3}} + \dfrac{D}{(t-a_2)^{1}} + \dfrac{E}{(t-a_2)^{2}} \\
+ \dfrac{E \cdot t + F}{(t^2 + B)^{1}} + \dfrac{G \cdot t + H}{(t^2 + B)^{2}},
$$
di conseguenza
$$
q(t) + A_1 + \dfrac{ a_1 \cdot A_1 + A_2}{(t-a_1)^{1}} + \dfrac{B_1}{(t-a_1)^{1}} + \dfrac{ B_1\cdot a_1 + B_2}{(t-a_1)^{2}} + \dfrac{C_1}{(t-a_1)^{2}} + \dfrac{ a_1 \cdot C_1 + C_2}{(t-a_1)^{3}} + \dfrac{D}{(t-a_2)^{1}} + \dfrac{E}{(t-a_2)^{2}} \\
+ \dfrac{E \cdot t + F}{(t^2 + B)^{1}} + \dfrac{G \cdot t + H}{(t^2 + B)^{2}}.
$$

Un argomento simile può adottarsi anche per le ultime due frazioni. Immaginate di aver quasi scomposto la frazione e di essere arrivati a

$$
q(t) + A_1 + \dfrac{ a_1 \cdot A_1 + A_2}{(t-a_1)^{1}} + \dfrac{B_1}{(t-a_1)^{1}} + \dfrac{ B_1\cdot a_1 + B_2}{(t-a_1)^{2}} + \dfrac{C_1}{(t-a_1)^{2}} + \dfrac{ a_1 \cdot C_1 + C_2}{(t-a_1)^{3}} + \dfrac{D}{(t-a_2)^{1}} + \dfrac{E}{(t-a_2)^{2}} \\
+ \dfrac{E_1 t^3 + E_2 t^2 + E_1 t + E_0}{(t^2 + B)^{1}} + \dfrac{G_1 t^3 + G_2 t^2 + G_1 t + G_0}{(t^2 + B)^{2}}.
$$

Effettuando un ultimo passo di divisione polinomiale si ottiene

$$
q(t) + A_1 + \dfrac{ a_1 \cdot A_1 + A_2}{(t-a_1)^{1}} + \dfrac{B_1}{(t-a_1)^{1}} + \dfrac{ B_1\cdot a_1 + B_2}{(t-a_1)^{2}} + \dfrac{C_1}{(t-a_1)^{2}} + \dfrac{ a_1 \cdot C_1 + C_2}{(t-a_1)^{3}} + \dfrac{D}{(t-a_2)^{1}} + \dfrac{E}{(t-a_2)^{2}} \\
+ \dfrac{ (t^2 + B) \cdot (F_3 t + F_2) + F_1 \cdot t + F_0}{(t^2 + B)^{1}} + \dfrac{ (H_3 t + H_2) \cdot (t^2 + B) \cdot t + H_1 \cdot t + H_0}{(t^2 + B)^{2}},
$$
ossia

$$
q(t) + A_1 + \dfrac{ a_1 \cdot A_1 + A_2}{(t-a_1)^{1}} + \dfrac{B_1}{(t-a_1)^{1}} + \dfrac{ B_1\cdot a_1 + B_2}{(t-a_1)^{2}} + \dfrac{C_1}{(t-a_1)^{2}} + \dfrac{ a_1 \cdot C_1 + C_2}{(t-a_1)^{3}} + \dfrac{D}{(t-a_2)^{1}} + \dfrac{E}{(t-a_2)^{2}} \\
+ F_3 t + F_2 + \dfrac{F_1 \cdot t + F_0}{(t^2 + B)^{1}} +  \dfrac{H_3 t + H_2}{t^2 + B} + \dfrac{ H_1 \cdot t + H_0}{(t^2 + B)^{2}}.
$$

**Osservazione finale:**  
Il polinomio $q(t)$ nella decomposizione sopra si ottiene tramite divisione polinomiale fra $p(t)$ e il denominatore.  
Infine, qualsiasi polinomio di secondo grado $t^2+ bt + c$ a delta negativo, ossia irriducibile, può essere scritto come somma di quadrati

$$
t^2 + b t + c = t^2 + bt + \dfrac{b^2}{4} + c - \dfrac{b^2}{4} = \left(t + \dfrac{b}{2}\right)^2 + \left(\sqrt{c - \dfrac{b^2}{4}}\right)^2.
$$

Quindi a meno di sostituire $t$ con $t + \dfrac{b}{2}$ ci si può ricondurre all'esempio precedente.

**NB:** L'esempio finale dovrebbe essere abbastanza generale per comprendere tutti i casi di decomposizione di funzioni razionali.  
Rimando a questo [link di LibreTexts](https://math.libretexts.org/Bookshelves/Calculus/Map%3A_Calculus__Early_Transcendentals_(Stewart)/07%3A_Techniques_of_Integration/7.04%3A_Integration_of_Rational_Functions_by_Partial_Fractions) per una trattazione generale.


### Esempio:

$$
\int \dfrac{2t }{(t+2)^{2}(t+1)} dt
$$

La decomposizione da ottenere è
$$
 \dfrac{A}{t+2} + \dfrac{B}{(t+2)^2}  + \dfrac{C}{t+1}
$$

Scrivendo la somma come un'unica funzione razionale si ottiene il seguente numeratore

$$
A(t+2)(t+1) + B(t+1) + C(t+2)^2 = 2t .
$$

Per calcolare i coefficienti $A$ $B$ e $C$ si può procedere in due modi

**Primo metodo:**
Svolgere il prodotto a sinistra
$$
A\cdot (t^2 + 3t +2)  + B\cdot (t + 1 ) + C \cdot (t^2 + 4t + 4) = (A + C) \cdot t^2 + (3A + B + 4C) \cdot t + 2\cdot A + B + 4C
$$
e risolvere il sistema

$$
\begin{cases} A + C = 0 \\ 3A + B + 4C = 2 \\ 2A + B + 4C = 0 \end{cases}
$$

ottenendo
$$
A = 2, \ B = 4, \ C = -2.
$$

Oppure osservare che i due polinomi devono coincidere $A(t+2)(t+1) + B(t+1) + C(t+2)^2$ e $2t$ devono coincidere in ogni punto quindi in particolare devono coincidere in $t = -2, -1$ e $0.$ Scegliere questi punti significa scegliere delle equazioni particolarmente facili da risolvere per trovare $A, B, C.$ Infatti,
per $t = -2$ otteniamo
$$
A \cdot 0 \cdot (-2 + 1) + B \cdot ( -2 + 1) + C \cdot 0 = - 2 \cdot 2
$$
e
$$
-B = -4 \rightarrow B = 4.
$$
Per $t = -1$ otteniamo
$$
A(-1+2)\cdot 0 + B \cdot 1 + C(-1+2)^2 = -2
$$
ossia
$$
C = -2.
$$
Infine, per $t=0$ si ottiene
$$
2\cdot 1 \cdot  A + 1 \cdot B + 4 C = 0,
$$
sostituendo i valori di $B$ e $C$ otteniamo
$$
2A + 4 - 4 \cdot 2 = 0 \Rightarrow A = 2.
$$
Trovata la decomposizione l'integrale è immediato

$$
\int \dfrac{2}{t+2} + \dfrac{4}{(t+2)^2} - \dfrac{2}{t+1} dt = 2 \ln(t+2) - 4 \dfrac{1}{t+2} + 2 \ln(2+t)
$$

### Sostituzione/ Antiderivazione + funzione composta:

$$
\int f'(x) \frac{1}{1 + f(x)^2} dx
$$


<details>
  <summary><strong>Mostra soluzione</strong></summary>
**Sostituzione diretta (con passaggi formali)**

1. **Sostituisci:**
   $$ y = f(x) $$

2. **Calcola il differenziale (passaggio formale):**
   La notazione $\frac{dy}{dx} = f'(x)$ suggerisce, ed è corretto, "moltiplicare" entrambi i membri per $dx$:
   $$ dy = f'(x)dx $$
   Questo è un *passaggio formale*: una manipolazione algebrica suggerita dalla notazione ma che è corretta in questo caso per il teorema di cambiamento di variabile.

3. **Sostituisci nell'integrale:**
   $$ \int \frac{1}{1 + f(x)^2} f'(x) dx = \int \frac{1}{1 + y^2} dy $$

4. **Risolvi l'integrale:**
   $$ \int \frac{1}{1 + y^2} dy = \arctan(y) + C $$

5. **Risostituisci:**
   $$ \arctan(y) + C = \arctan(f(x)) + C $$

#### 2. Ricavare x e dx (meno efficiente)

1. **Ricava l'inversa:**
   $$ x = f^{-1}(y) $$

2. **Calcola il differenziale:**
   $$ dx = \frac{df^{-1}}{dy}(y) \cdot dy = \frac{1}{f'(f^{-1}(y))} dy $$

3. **Sostituisci nell'integrale:**
   $$ \int f'(y) \cdot \frac{1}{1 + y^2} \frac{1}{f'(y)} dy = \int \frac{1}{1 + y^2} dy $$

4. **Risolvi l'integrale:**
   $$ \int \frac{1}{1 + y^2} dy = \arctan(y) + C $$

5. **Risostituisci:**
   $$ \arctan(y) + C = \arctan(f(x)) + C $$

#### 3. Verifica con la derivata (controllo)

1. **Ricorda la primitiva:**
   La primitiva di $\frac{1}{1 + y^2}$ è $\arctan(y)$.

2. **Verifica con la derivata:**
   Deriva $\arctan(f(x))$ per controllare se ottieni l'integranda originale.

### Esercizio 2


$$
\int x \cdot e^{-x^2} dx
$$

provare a fare con sostituzione



<details>
  <summary><strong>Mostra soluzione</strong></summary>
  $$
  \begin{gather*}
  x^2 = t \Rightarrow x = \sqrt{t} \quad \text{abbiamo scelto il ramo positivo} \\
  dx = \dfrac{1}{2\sqrt{t}} \ dt \\
  \int \sqrt{t} \cdot e^{-t} \frac{1}{2 \sqrt{t}} \ dt \\
  = \frac{1}{2} \int e^{-t} dt = -\frac{1}{2} e^{-t} + C = -\frac{1}{2} e^{-x^2} + C 
  \end{gather*}
  $$

</details>


# Integrazione per parti, i trucchi più comuni

Le descrizioni che seguono non sono rigorose sono solo una indicazione su cosa cercare e per fare pratica e acquisire quei due o tre trucchi che servono.

- **Integrare 1 o una potenza di t:**
  $$\int t^3 ln(t) $$
  $$ \int t \cdot arctan(t) dt$$

<details>
  <summary><strong>Mostra soluzione</strong></summary>
  $$
    \begin{gather*}
    \dfrac{t^4}{4} - \int \frac{t^4}{4} \cdot \dfrac{1}{t} \ dt \\
    = \dfrac{t^4}{4} - \int \frac{t^4}{4} \cdot \dfrac{1}{t} \ dt \\
    = \dfrac{t^4}{4} - \frac{t^4}{16} + C 
    \end{gather*}
  $$

  $$
    \begin{gather*}
    \dfrac{t^2}{2\cdot ( 1 + t^2)} - \int \frac{t^2}{2\cdot (1 + t^2)} \ dt \\
    = \dfrac{t^2}{2\cdot ( 1 + t^2)} - \int \frac{t^2 + 1 - 1 }{2\cdot (1 + t^2)} \ dt \\
    = \dfrac{t^2}{2\cdot ( 1 + t^2)} - \dfrac{t}{2} + \int \frac{ 1 }{2\cdot (1 + t^2)} \ dt \\
    = \dfrac{t^2}{2\cdot ( 1 + t^2)} - \dfrac{t}{2} + \frac{ \arctan(t)}{2} + C
    \end{gather*}
  $$

</details>



  
- **Polinomio contro esponenziale o funzione trigonometrica:**
  Idea integrare continuamente quella con la primitiva più facile per abbassare di grado il polinomio fino a ottenere l'integrale facile

  $$ \int t^3 e^{-3t} dt$$

<details>
<summary><strong>Mostra soluzione</strong></summary>

$$
    \begin{gather*}
    \int t^3 e^{-3t}dt \\
    \quad \text{Sia } u=t^3,\; dv=e^{-3t}dt \quad \Rightarrow \quad du=3t^2dt,\; v=-\frac{1}{3}e^{-3t} \\
    \int t^3e^{-3t}dt = -\frac{1}{3}t^3e^{-3t}+\int t^2e^{-3t}dt \\[1mm]
    \quad \text{Per } \int t^2e^{-3t}dt,\; \text{scegliamo } u=t^2,\; dv=e^{-3t}dt \\
    \quad \Rightarrow \quad du=2tdt,\; v=-\frac{1}{3}e^{-3t} \\
    \int t^2e^{-3t}dt = -\frac{1}{3}t^2e^{-3t}+\frac{2}{3}\int te^{-3t}dt \\[1mm]
    \quad \text{Per } \int te^{-3t}dt,\; poniamo } u=t,\; dv=e^{-3t}dt \\
    \quad \Rightarrow \quad du=dt,\; v=-\frac{1}{3}e^{-3t} \\
    \int te^{-3t}dt = -\frac{1}{3}te^{-3t}+\frac{1}{3}\int e^{-3t}dt \\
    \quad = -\frac{1}{3}te^{-3t}-\frac{1}{9}e^{-3t} \\[1mm]
    \text{Quindi, } \int t^2e^{-3t}dt = -\frac{1}{3}t^2e^{-3t}-\frac{2}{9}te^{-3t}-\frac{2}{27}e^{-3t} \\[1mm]
    \text{e infine: } \int t^3e^{-3t}dt = -\frac{1}{3}t^3e^{-3t}-\frac{1}{3}t^2e^{-3t}-\frac{2}{9}te^{-3t}-\frac{2}{27}e^{-3t}+C
    \end{gather*}
$$

</details>
