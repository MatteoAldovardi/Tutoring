### Invertire graficamente una funzione:

Data una funzione invertibile $ y = f(x)$ e il suo grafico 
$$\text{Grafico}(f) = \{ (x, f(x)): x \in \text{dominio}(f) \}$$
Il grafico della sua inversa si ottiene specchiando la funzione rispetto alla retta $y = x.$ Equivalentemente, il grafico di $f^{-1}$ si ottiene applicando questa trasformazione 
$$ 
\begin{cases} y = x \\ x = y \end{cases} \quad ,
$$
ovvero scambiando $x$ e $y.$
Una idea del perché si può avere riscrivendo un generico punto del grafico di $f$:

$$(x, f(x)) = ( f^{-1}\left(f(x)\right), f(x))$$ e chiamando $t = f(x)$ otteniamo
$$(x,f(x))=(f^{-1}(t), t)$$. Adesso quello che manca per ottenere il punto di coordinate $(t,f^{-1}(t))$ è scambiare le due coordinate. 


Per convincersi che la trasformazione sopra di scambio delle coordinate consiste nello specchiare i punti rispetto alla retta $y=x$ consiglio di farsi una paio di esempi in cui si prova a disegnare il grafico di qualche funzione nota $y=e^{x}, \ y = x^3, \dots$ e a invertirla graficamente. 



### Esempio:

Grafico di $y=e^x$ e della sua inversa

![Inversione dell'esponenziale](immagini/inversione_exp.png)

```python
# Di sotto c'è del codice che puoi eseguire per invertire altre funzioni 


import numpy as np
import matplotlib.pyplot as plt

def invertitore_grafico(f, x_range, num_points=100, label="f(x)", save=False, path="immagini"):
    # Generate values for x and corresponding y values from the function f
    x = np.linspace(*x_range, num_points)
    y = f(x)
    
    # Select a random point from the function
    idx = np.random.randint(0, num_points)
    x_sample, y_sample = x[idx], y[idx]
    
    # Create the plot
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.plot(x, y, label=label, color='blue')  # Original function
    ax.plot(y, x, label=f'Inversa di {label}', color='red', linestyle='dashed')  # Inverse function
    
    ax.plot(x, x, color='gray', linestyle='dotted', label='y = x')  # Line y = x
    
    # Draw the line connecting the point and its transformed counterpart
    ax.plot([x_sample, y_sample], [y_sample, x_sample], color='green', linestyle='dotted', linewidth=2)
    ax.scatter([x_sample, y_sample], [y_sample, x_sample], color='black')  # Highlight the points
    
    # Set plot labels and title
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.legend()
    ax.set_title("Invertitore Grafico")
    plt.grid(True)
    
    # Save the plot if requested
    if save:
        plt.savefig(path)
    plt.show()
    

# Example with an exponential function
invertitore_grafico(np.exp, (-2, 2), label="e^x")#, save=False, path="immagini/inversione_exp.png")
invertitore_grafico(np.cos, (0, np.pi), label="cos(x)")
```

## Funzioni Pari

Una funzione $f(x)$ è detta **pari** se, per ogni $x$ nel suo dominio, vale la relazione:

$f(x) = f(-x)$

Questa proprietà implica una simmetria speculare del grafico della funzione rispetto all'asse delle ordinate (l'asse y). In altre parole, se pieghiamo il grafico lungo l'asse y, le due metà coincidono perfettamente.

### Esempi di Funzioni Pari

* **Funzioni Potenza con Esponente Pari**:
    * $f(x) = x^2$
    * $f(x) = x^4$
    * $f(x) = x^6$
* **Funzioni Trigonometriche**:
    * $f(x) = \cos(x)$
* **Funzioni con Valore Assoluto**:
    * $f(x) = |x|$
    * $f(x) = x^2 + |x|$
* **Funzioni a cui non so assegnare un nome in particolare**:
    * $f(x) = \dfrac{\sin(x)}{x}$

### Vantaggi dello Studio delle Funzioni Pari

La simmetria delle funzioni pari semplifica notevolmente il loro studio. Invece di analizzare l'intero dominio, possiamo concentrarci solo sulla metà positiva o negativa. Una volta compreso il comportamento della funzione in una di queste metà, possiamo facilmente dedurre l'altra metà per simmetria.

### Esempio Grafico: $f(x) = x^2$

Consideriamo la funzione $f(x) = x^2$. Per verificare che sia pari, calcoliamo $f(-x)$:

$f(-x) = (-x)^2 = x^2 = f(x)$

Poiché $f(-x) = f(x)$, la funzione è pari.

## Funzioni Dispari

Una funzione $f(x)$ è detta **dispari** se, per ogni $x$ nel suo dominio, vale la relazione:

$f(-x) = -f(x)$

Questa proprietà implica una simmetria rotazionale di 180 gradi del grafico della funzione rispetto all'origine (il punto (0, 0)). In altre parole, se ruotiamo il grafico di 180 gradi attorno all'origine, il grafico risultante coincide con il grafico originale.

### Esempi di Funzioni Dispari

* **Funzioni Potenza con Esponente Dispari**:
    * $f(x) = x$
    * $f(x) = x^3$
    * $f(x) = x^5$
* **Funzioni Trigonometriche**:
    * $f(x) = \sin(x)$
    * $f(x) = \tan(x)$
* **Altre Funzioni**:
    * $f(x) = 1/x$
    * $f(x) = x\cdot \cos(x)$

## Caratteristiche principali

* Il grafico di una funzione dispari passa sempre per l'origine (0, 0), a meno che non sia esclusa dal dominio.
* Se conosciamo il comportamento di una funzione dispari per $x > 0$, possiamo facilmente dedurre il comportamento per $x < 0$ cambiando il segno dei valori della funzione.

## Come Determinare se una Funzione è Dispari

1.  **Calcola $f(-x)$**: Sostituisci $-x$ al posto di $x$ nella funzione.
2.  **Confronta $f(-x)$ con $-f(x)$**:
    * Se $f(-x) = -f(x)$, la funzione è dispari.
    * Se $f(-x) = f(x)$, la funzione è pari.
    * Se nessuna delle due condizioni è soddisfatta, la funzione non è né pari né dispari.

## Esempio

Prendiamo la funzione $f(x) = x^3$.

* Calcoliamo $f(-x)$: $f(-x) = (-x)^3 = -x^3$
* Confrontiamo con $-f(x)$: $-f(x) = -x^3$

Poiché $f(-x) = -f(x)$, la funzione $f(x) = x^3$ è dispari.

## Traslazione dei Grafici

La traslazione di un grafico consiste nello spostare l'intera curva lungo l'asse x o l'asse y, senza alterarne la forma.

### Traslazione Lungo l'Asse Y

* Per traslare il grafico di $f(x)$ verso l'alto di $c$ unità, aggiungiamo $c$ alla funzione: $g(x) = f(x) + c$.
* Per traslare il grafico di $f(x)$ verso il basso di $c$ unità, sottraiamo $c$ alla funzione: $g(x) = f(x) - c$.

**Esempio:**

* Consideriamo $f(x) = x^2$.
* Per traslarlo verso l'alto di 2 unità, otteniamo $g(x) = x^2 + 2$.
* Per traslarlo verso il basso di 3 unità, otteniamo $h(x) = x^2 - 3$.


![Traslazione](immagini/traslazioney.png)


### Altro modo di vedere questa traslazione

Per riuscire a disegnare il grafico di $y = f(x) + c$ si può pensare di adottare un nuovo sistema di assi $x', y'$ la cui origine è localizzata in $(0,c).$
La trasformazione delle coordinate per questo nuovo sistema di assi è:

$$
\begin{gather*}
x' = x \\
y' = y - c
\end{gather*}
$$

In questo nuovo sistema di assi, l'equazione della funzione ridiventa $y' = f(x').$
Infatti, si può riscrivere come $y - c = f(x)$ e ponendo
$y'= y - c$ e $x' = x$ otteniamo $y' = f(x').$
Quindi per disegnare il grafico $y = f(x) + c$ disegnado quello di $f$ usando i nuovi assi $x'$ e $y'.$

## Traslazione Lungo l'Asse X

La traslazione di un grafico lungo l'asse x consiste nello spostare l'intera curva orizzontalmente, senza alterarne la forma.

### Regole Fondamentali

* **Traslazione verso destra**: Per traslare il grafico di $f(x)$ verso destra di $c$ unità, sostituiamo $x$ con $(x - c)$ nella funzione: $g(x) = f(x - c)$.
* **Traslazione verso sinistra**: Per traslare il grafico di $f(x)$ verso sinistra di $c$ unità, sostituiamo $x$ con $(x + c)$ nella funzione: $g(x) = f(x + c)$.

### Esempio Base: $f(x) = x^2$

* Traslazione verso destra di 2 unità: $g(x) = (x - 2)^2$. Il vertice della parabola si sposta da (0, 0) a (2, 0).
* Traslazione verso sinistra di 3 unità: $h(x) = (x + 3)^2$. Il vertice della parabola si sposta da (0, 0) a (-3, 0).

### Altri Esempi

1.  **$y = \sin(x)$**:
    * Traslazione verso destra di $\pi/2$ unità: $y = \sin(x - \pi/2) = -\cos(x)$.
    * Traslazione verso sinistra di $\pi/2$ unità: $y = \sin(x + \pi/2) = \cos(x)$.
2.  **$y = |x|$**:
    * Traslazione verso destra di 1 unità: $y = |x - 1|$.
    * Traslazione verso sinistra di 1 unità: $y = |x + 1|$.
3.  **$y = 1/x$**:
    * Traslazione verso destra di 2 unitá: $y = 1/(x-2)$
    * Traslazione verso sinistra di 2 unitá: $y = 1/(x+2)$

### Altro Modo di Vedere la Traslazione Orizzontale

Proprio come per le traslazioni verticali, possiamo interpretare le traslazioni orizzontali utilizzando un nuovo sistema di coordinate.

Per disegnare il grafico di $y = f(x - c)$, possiamo pensare di adottare un nuovo sistema di assi $x', y'$ la cui origine è localizzata in $(c, 0)$.

La trasformazione delle coordinate per questo nuovo sistema di assi è:

$$
\begin{gather*}
x' = x - c \\
y' = y
\end{gather*}
$$

In questo nuovo sistema di assi, l'equazione della funzione ridiventa $y' = f(x')$.

Infatti, si può riscrivere come $y = f(x')$, e ponendo $x' = x - c$ e $y' = y$ otteniamo $y' = f(x')$. Quindi, in maniera simile a prima, è sufficiente disegnare il grafico di $f$ usando gli assi $x'$ e $y'.$


## Riscalatura dei Grafici

La riscalatura di un grafico consiste nel modificare la "dimensione" della curva lungo l'asse $ x $ o l'asse $ y $.

### Riscalatura Lungo l'Asse Y

* Per allungare il grafico di $ f(x) $ verticalmente di un fattore $ c $ con $ c > 1 $, moltiplichiamo la funzione per $ c $:  
  $$
  g(x) = c \cdot f(x).
  $$
* Per comprimere il grafico di $ f(x) $ verticalmente di un fattore $ c $ (con $ 0 < c < 1 $), moltiplichiamo la funzione per $ c $:  
  $$
  g(x) = c \cdot f(x).
  $$

**Esempio:**

* Consideriamo $ f(x) = \sin(x) $.
* Per allungarlo verticalmente di un fattore $2$, otteniamo $ g(x) = 2 \sin(x) $.
* Per comprimerlo verticalmente di un fattore $0.5$, otteniamo $ h(x) = 0.5 \sin(x) $.

### Interpretazione Alternativa della Riscalatura Lungo $ Y $

Come nei casi precedenti, possiamo ricondurci di nuovo al grafico $ y' = f(x') $ adottando un nuovo sistema di coordinate. Infatti, applicando

$$
\begin{gather*} 
y ' = \frac{y}{c} \\
x' = x 
\end{gather*}
$$

a $ y = c \cdot f(x) $, otteniamo:

$$
\frac{y}{c} = f(x) 
$$

e quindi 

$
y' = f(x').
$


In conclusione, il grafico si può ottenere disegnando il grafico di $f$ e cambiando la scala lungo $y$ di un fattore $\frac{1}{c}.$



### Riscalatura Lungo l'Asse $X$

* Per allungare il grafico di $ f(x) $ orizzontalmente di un fattore $ c $ (con $ c > 1 $), sostituiamo $ x $ con $ \frac{x}{c} $:
  $$
  g(x) = f \left( \frac{x}{c} \right).
  $$
* Per comprimere il grafico di $ f(x) $ orizzontalmente di un fattore $ c $ (con $ 0 < c < 1 $), sostituiamo $ x $ con $ \frac{x}{c}$:
  $$
  g(x) = f \left( \frac{x}{c} \right).
  $$

**Esempio:**

* Consideriamo $ f(x) = \sin(x) $.
* Per allungarlo orizzontalmente di un fattore 2, otteniamo $ g(x) = \sin \left( \frac{x}{2} \right) $.
* Per comprimerlo orizzontalmente di un fattore 0.5, otteniamo $ h(x) = \sin(2x) $.

### Interpretazione Alternativa della Riscalatura Lungo $ X $

Analogamente al caso precedente, possiamo interpretare la trasformazione come un cambio di variabili:

$$
\begin{gather*} 
y' = y \\
x' = \frac{x}{c} 
\end{gather*}
$$

Sostituendo in $ y = f(\frac{x}{c}) $, otteniamo:

$$
y' = f(x').
$$


In conclusione, il grafico si può ottenere disegnando il grafico di $f$ e cambiando la scala lungo $x$ di un fattore $\frac{1}{c}.$
