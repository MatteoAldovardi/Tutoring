# Syllabus for an Introductory Statistics Course  
**Based on Textbook Content**  

---

## Course Overview  
This course introduces fundamental statistical concepts and methods, covering data collection, descriptive statistics, probability, probability distributions, hypothesis testing, and inferential statistics.  

---

## Course Structure  
- **Duration:** 15 weeks (including exams and review).  
- **Textbook:** Chapters as per provided content.  

---

## Weekly Breakdown  

### Week 1-2: Introduction to Statistics and Data Collection  
- **Topics:**  
  - Definition of statistics and the statistical process.  
  - Types of data: qualitative vs. quantitative, discrete vs. continuous.  
  - Sampling methods (probabilistic and non-probabilistic).  
  - Observational studies vs. experiments.  
  - Sources of sampling error.  
- **Readings:**  
  - Chapter 1 (Sections 1.1–1.5).  

---

### Week 3-4: Descriptive Statistics  
- **Topics:**  
  - Organizing qualitative/quantitative data (tables, bar charts, pie charts).  
  - Measures of central tendency (mean, median, mode).  
  - Measures of dispersion (range, variance, standard deviation).  
  - Five-number summary, boxplots, z-scores, and outliers.  
- **Readings:**  
  - Chapter 3 (Sections 3.1–3.5).  

---

### Week 5-6: Bivariate Data Analysis  
- **Topics:**  
  - Scatterplots, correlation coefficients, and linear regression.  
  - Coefficient of determination (\(R^2\)).  
  - Contingency tables and association analysis.  
- **Readings:**  
  - Chapter 4 (Sections 4.1–4.4).  

---

### Week 7-8: Probability Fundamentals  
- **Topics:**  
  - Probability rules (empirical, classical, simulations).  
  - Additive/multiplicative rules, complementary events.  
  - Independent and conditional events.  
- **Readings:**  
  - Chapter 5 (Sections 5.1–5.3).  

---

### Week 9-10: Discrete and Normal Distributions  
- **Topics:**  
  - Discrete variables: mean, variance, binomial distribution.  
  - Normal distribution: properties, z-scores, applications.  
  - Normal approximation to binomial.  
- **Readings:**  
  - Chapters 6 and 7.  

---

### Week 11-12: Hypothesis Testing  
- **Topics:**  
  - Null/alternative hypotheses, Type I/II errors.  
  - Hypothesis tests for means (σ known/unknown) and proportions.  
  - Interpretation of p-values.  
- **Readings:**  
  - Chapter 10 (Sections 10.1–10.5).  

---

### Week 13: Chi-Square Tests  
- **Topics:**  
  - Goodness-of-fit and independence tests.  
- **Readings:**  
  - Chapter 11 (Sections 11.1–11.2).  

---

### Week 14: Review and Applications  
- **Topics:**  
  - Case studies integrating statistical methods.  
  - Ethical considerations in reporting.  

---

### Week 15: Final Exam and Project Presentations  
- **Final Project:**  
  - Statistical analysis of a real-world dataset.  

---

## Assessment  
- **Homework:** 30%  
- **Midterm Exam:** 25%  
- **Final Project:** 25%  
- **Final Exam:** 20%
