```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Data from the table
data = {
    "Media inferiore non terminata": [10179, 945, 13271],  # Impiegato, Disoccupato, Escluso dalla forza lavoro
    "Media inferiore terminata": [33624, 2012, 25806],
    "Media superiore": [35407, 1823, 17089],
    "Laurea": [49534, 1615, 17415],
}

# Compute total counts for each education level
total_med_inf_nt = sum(data["Media inferiore non terminata"])
total_med_inf_t = sum(data["Media inferiore terminata"])
total_med_sup = sum(data["Media superiore"])
total_laur = sum(data["Laurea"])

# Compute conditional probabilities explicitly
p_imp_med_inf_nt = data["Media inferiore non terminata"][0] / total_med_inf_nt
p_imp_med_inf_t = data["Media inferiore terminata"][0] / total_med_inf_t
p_imp_med_sup = data["Media superiore"][0] / total_med_sup
p_imp_laur = data["Laurea"][0] / total_laur

p_dis_med_inf_nt = data["Media inferiore non terminata"][1] / total_med_inf_nt
p_dis_med_inf_t = data["Media inferiore terminata"][1] / total_med_inf_t
p_dis_med_sup = data["Media superiore"][1] / total_med_sup
p_dis_laur = data["Laurea"][1] / total_laur

p_excl_med_inf_nt = data["Media inferiore non terminata"][2] / total_med_inf_nt
p_excl_med_inf_t = data["Media inferiore terminata"][2] / total_med_inf_t
p_excl_med_sup = data["Media superiore"][2] / total_med_sup
p_excl_laur = data["Laurea"][2] / total_laur

# Store probabilities in a structured form
conditional_probs = {
    "Media inferiore non terminata": [p_imp_med_inf_nt, p_dis_med_inf_nt, p_excl_med_inf_nt],
    "Media inferiore terminata": [p_imp_med_inf_t, p_dis_med_inf_t, p_excl_med_inf_t],
    "Media superiore": [p_imp_med_sup, p_dis_med_sup, p_excl_med_sup],
    "Laurea": [p_imp_laur, p_dis_laur, p_excl_laur],
}

# Convert to DataFrame
df_conditional = pd.DataFrame(conditional_probs, index=["Impiegato", "Disoccupato", "Escluso dalla forza lavoro"])

# Print probabilities for verification
print(df_conditional)

# Plot conditional histogram
fig, ax = plt.subplots(figsize=(8, 5))
df_conditional.T.plot(kind="bar", stacked=True, colormap="viridis", ax=ax)

# Formatting
plt.title("Distribuzione condizionata dello stato occupazionale dato il titolo di studio")
plt.xlabel("Titolo di studio")
plt.ylabel("Frequenza relativa")
plt.xticks(rotation=45)
plt.legend(title="Stato occupazionale")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Data from the table
data = {
    "Media inferiore non terminata": [10179, 945, 13271],  # Impiegato, Disoccupato, Escluso dalla forza lavoro
    "Media inferiore terminata": [33624, 2012, 25806],
    "Media superiore": [35407, 1823, 29014],
    "Laurea": [49534, 1615, 73581],
}

# Convert to DataFrame
df = pd.DataFrame(data, index=["Impiegato", "Disoccupato", "Escluso dalla forza lavoro"])

# Compute conditional probabilities
df_conditional = df.div(df.sum(axis=0), axis=1)

# Plot conditional histogram
fig, ax = plt.subplots(figsize=(8, 5))
df_conditional.T.plot(kind="bar", stacked=True, colormap="viridis", ax=ax)

# Formatting
plt.title("Distribuzione condizionata dello stato occupazionale dato il titolo di studio")
plt.xlabel("Titolo di studio")
plt.ylabel("Frequenza relativa")
plt.xticks(rotation=45)
plt.legend(title="Stato occupazionale")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show plot
plt.tight_layout()
plt.show()
```
