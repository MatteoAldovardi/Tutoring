# Tipi di indagini statistiche, tipi di variabili e tecniche di campionamento.

## Studio osservazionale:
Misura i valori della variabile risposta senza provare ad influenzare il risultato 
Esempio: Studi osservazionali sul legame tra cancro ai polmoni e fumare sigarette.
Negli studi osservazionali non si può dimostrare un legame di causalità tra la variabili esplicative e la variabile risposta, ma solamente un'associazione. 

## Confusione:
Quando gli effetti di due o più variabili espicative non sono separati. 


## Variabile nascosta:
È una variabile esplicativa che non è stata considerata, ma che influenza la risposta.

## Variabile confondente:
Una variabile confodente è una variabile che influenza sia le variabili dipendenti che indipendenti creando associazioni spurie che non rendono possibile formulare chiaramente una relazione causa effetto.

## Studio osservazionale trasversale:
Raccolta dati su individui in un particolare momento storico o intervallo di tempo breve. 
È un tipo di studio economico, ma con dei limiti. Per esempio nello studio sul cancro ai polmoni si può considerare un gruppo di persone dividerle in fumatrici e non fumatrici confrontando l'incidenza del cancro ai polmoni. Limiti del metodo : non si tiene conto degli individui che svilupperanno la patologia dopo la raccolta dati.


## Studi sul controllo dei casi:
Studi retrospettivi, di indagine del passato o attraverso intervista o attaverso il controllo di dati già esistenti.
In questi studi si possono *abbinare* individui che presentano o non presentano una certa caratteristica. 
Abbinare significa scegliere individui il più possibile omogenei in termini demografici e altre variabili che possono influenzare la variabile risposta. 
Nel caso del fumo, dopo aver abbinato gli individui possiamo chiedere agli individui di ciascun gruppo quanto hanno fumato negli ultimi $25$ anni. 

Uno svantaggio è sicuramente l'affidabilità parziale delle dichiarazioni degli intervistati.
Alcuni vantaggi sono la relativa rapidità ed economicità. 

## Studi di coorte
Uno studio di coorte identifica un gruppo di individui esposti allo stesso fenomeno (la coorte) e li segue nel tempo. Alcuni individui saranno esposti a determinati fattori, mentre altri no. Alla fine dello studio, viene registrato il valore della variabile di risposta.

Gli studi di coorte richiedono lunghi periodi di osservazione e sono quindi prospettici. Inoltre, necessitano di un ampio numero di partecipanti e presentano il rischio di abbandono da parte di alcuni individui. Nonostante questi svantaggi, rappresentano il tipo di studio osservazionale più potente.

**Esempio famoso: Birth to Twenty**
Il progetto Birth to Twenty, condotto in Sudafrica dal 1990 al 2010, ha seguito 3.273 bambini dalla nascita fino ai 20 anni. Lo studio è iniziato nel 1989/1990 con interviste alle madri in gravidanza per raccogliere dati demografici e informazioni sulle condizioni della gravidanza. Dopo valutazioni alla nascita e a sei mesi, la raccolta dati è proseguita annualmente.

I sondaggi hanno coperto aspetti come salute, sviluppo, ambiente, istruzione, nutrizione e status socioeconomico del bambino, oltre a opinioni politiche e sociali. Nel tempo, lo studio è stato una fonte preziosa di informazioni sulla generazione nata nel 1990.

Ad esempio, intorno al 1997, i partecipanti della coorte sono stati intervistati sulla loro conoscenza dei brand di sigarette. Il fatto che molti bambini riconoscessero diversi marchi ha spinto il Ministero della Salute a vietare la pubblicità delle sigarette e a proibirne la vendita ai minori( fonte Wikipedia: [Birth to Twenty](https://en.wikipedia.org/wiki/Birth_to_Twenty).)



## Campionamento casuale:
Per decidere gli individui da includere nel campione si usa un procedimento casuale.

## Campionamento casuale semplice:
Si dispone di un elenco della popolazione di $N$ individui (frame). Da questo frame si estraggono $n$ in modo uniforme e senza reinmissione: un membro della popolazione non può essere scelto più volte


## Campione rappresentativo:
Un campione è rappresentativo quando riflette in piccolo la popolazione da cui è estratto. È più facile pensare a esempi in cui un campione è evidentemente non rappresentativo. Un caso famoso è il sondaggio del Literary Digest, che nel 1936 predisse erroneamente la vittoria di Landon contro Roosevelt alle elezioni presidenziali statunitensi. L'indagine era stata condotta tramite interviste telefoniche, escludendo così chi non possedeva un telefono, ovvero le fasce meno abbienti della popolazione. Poiché i possessori di telefoni erano in prevalenza persone benestanti, più inclini a votare per il repubblicano Landon, il campione risultò distorto e non rappresentativo dell'elettorato complessivo.

## Campione stratificato
Un campione stratificato si ottiene suddividendo la popolazione in gruppi non sovrapponibili, detti strati, che condividono alcune caratteristiche comuni. Successivamente, all'interno di ciascuno strato viene effettuato un campionamento casuale semplice. Il campione finale è quindi composto dall'unione dei campioni estratti da ogni strato. Un campionamento stratificato permette di ottenere le stesse informazioni o maggiore informazioni rispetto ad un campionamento semplice. Questo avviene perché all'interno di ogni strato di popolazione c'è più probabilità di ottenere la stessa risposta (**non troppo capito**).Infine, un campionamento stratificato assicura che ogni strato sia rappresentato, per definizione.  


## Campione non probabilistico
Un campione non probabilistico non viene selezionato con un procedimento casuale, come ad esempio estrarre seguendo distribuzione uniforme. Esistono diverse tecniche di campionamento non probabilistico, ma tra le più comuni vi sono il campione autoselezionato e il campione a risposta volontaria, come i sondaggi aperti durante una live streaming.
Questo tipo di sondaggi è poco affidabile, poiché negli ambienti online tendono a partecipare soprattutto persone con opinioni estreme, distorcendo i risultati. In generale, l'autoselezione è una distorsione. 

## Campione Sistematico
**Procedura con Frame (Lista della Popolazione)**
- Si fissa un numero k.
- Si estrae casualmente un numero L fra 1 e k.
- Per estrarre il campione, si seleziona l'individuo L-esimo della lista e poi si     procede di k in k finché non si ottiene un campione di n individui.

**Procedura Senza Frame**
 - Esempio: selezione di n clienti di un negozio.
 - Si fissa k e si estrae L come nel caso con frame.
 - Si seleziona il cliente L-esimo e poi si attende il cliente L+k-esimo, e così via, fino all'L+k⋅(n−1)-esimo cliente.
 - Vantaggio: non richiede sforzi preliminari per ottenere un frame.

**Scelta di $k$**
- Influenza la rappresentatività del campione.
- k non deve essere troppo piccolo:
    - Rischio di campione poco rappresentativo.
    - Nell'esempio del negozio, rischio di selezionare clienti solo in un numero ristretto di fasce orarie, introducendo una distorsione.
**k troppo grande:**
- Esempio: $k=30$ per un campione di $40$ persone si possono dover apettare sino a $1175$ individui e questo potrebbe non essere possibile.


## Campione a grappolo
Un campione a grappolo è ottenuto selezionando tutti gli individui all'interno di un gruppo (cluster) individuato casualmente.



## Campione multistadio
Per estrarre il campione si combinano più delle tecniche viste in precedenza.
Esempio un sondaggio di Census Bureau può arrivare a coinvolgere cinque stadi di campionamento:
**Fase 1:** campione stratificato
**Fase 2:** campione a grappolo
**Fase 3:** campione stratificato
**Fase 4:** campione a grappolo 
**Fase 5:** campione sistematico



## Taglia del campione:
Come sempre vale il principio del no free lunch. Campioni più grandi in generale garantiscono risultati più accurati, ma sono più costosi. Ogni indagine statistica deve trovare il giusto compromesso fra affidabilità dei risultati e costi.



## Errore di campionament

```python
import itertools
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# Campionamento casuale tramite la funzione numpy.random.choice 

popolazione = ["Bice", "Astolfo", "Giampiero", "Medardo", "Rodomonte"]

# Listare tutti i possibili campioni da 2 (senza considerare l'ordine)
possibili_campioni = list(itertools.combinations(popolazione, 2))
print("Tutti i possibili campioni di due individui:")
print(possibili_campioni)

# Verifica empirica che tutti i 10 campioni hanno la stessa probabilità di venire estratti 0.10
# Fai una estrazione random con numpy di due individui (senza considerare l'ordine)
np.random.seed(42)  # Per riproducibilità
n = 1000 # Numero di campioni considerati, testare anche 10^5
#Replace = false impedisce il reinserimento
due_individui = tuple(sorted(np.random.choice(popolazione, 2, replace=False))) 
print("Campione casuale di due individui:", due_individui)

# Ripeti per n campioni e conta le frequenze (senza considerare l'ordine)
campioni = [tuple(sorted(np.random.choice(popolazione, 2, replace=False))) for _ in range(n)]
conteggio = Counter(campioni)

# Normalizzazione delle frequenze
frequenze = {k: v / n for k, v in conteggio.items()}

# Grafico a barre delle frequenze
fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(range(len(frequenze)), frequenze.values(), tick_label=[str(k) for k in frequenze.keys()])
ax.set_xlabel("Campioni")
ax.set_ylabel("Frequenza relativa")
ax.set_title("Frequenze relative dei campioni di due individui (senza ordine)")
ax.set_xticks(range(len(frequenze)))
ax.set_xticklabels([str(k) for k in frequenze.keys()], rotation=45, ha="right")
plt.tight_layout()
plt.show()

# Commento sul risultato
print("Le barre hanno altezza circa 0.10 (probabilità 1/10), indicando una distribuzione uniforme dei 10 possibili campioni di due individui.")
```

```python
import random
import matplotlib.pyplot as plt

def campionamento_sistematico(popolazione, k, n):
    """
    Esegue il campionamento sistematico e restituisce il campione.

    Args:
        popolazione (list): La lista che rappresenta la popolazione.
        k (int): L'intervallo di campionamento.
        n (int): La dimensione del campione desiderato.

    Returns:
        list: Il campione estratto.
    """

    L = random.randint(1, k)
    campione = []
    indice_corrente = L - 1  # Gli indici in Python partono da 0

    while len(campione) < n:
        campione.append(popolazione[indice_corrente])
        indice_corrente += k
        if indice_corrente >= len(popolazione):
            break

    return campione

# Esempio di utilizzo
N = 100  # Dimensione della popolazione
popolazione = list(range(1, N + 1))
k = 10
n = 8

campione = campionamento_sistematico(popolazione, k, n)

# Visualizzazione
plt.figure(figsize=(12, 4))
plt.scatter(popolazione, [0] * len(popolazione), color='blue', alpha=0.5, label='Popolazione')
plt.scatter(campione, [0] * len(campione), color='red', s=50, label='Campione')
plt.yticks([])  # Rimuove i segni sull'asse y
plt.xlabel('Indice nella popolazione')
plt.title('Campionamento Sistematico')
plt.legend()
plt.grid(axis='x')
plt.show()

print("Campione estratto:", campione)
```
